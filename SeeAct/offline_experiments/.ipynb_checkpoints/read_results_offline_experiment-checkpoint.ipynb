### How to determine the acc in offline exps?

```python
for query_id, query in enumerate(query_meta_data_gpt4):
    assert len(query['pos_candidates']) == 1
    gt_node_id = query['pos_candidates'][0]['backend_node_id']
    pred = predictions_gpt4[query_id]
    print("----------------------------")
    print(query_id, gt_node_id, gt_node_id in [choice[0] for choice in query['choices']])
    for c in query['choices']:
        print(c)
    print(pred['prompt'][2])
    print(pred['gpt_output'][0])
    print(pred['gpt_output'][1])
```

```python
for query_id, query in enumerate(query_meta_data_gpt4):
    assert len(query['pos_candidates']) == 1
    gt_node_id = query['pos_candidates'][0]['backend_node_id']
    if gt_node_id in [choice[0] for choice in query['choices']]:
        pred = predictions_gpt4[query_id]
        print(query_id, gt_node_id)
        print("------------------------------------------------------------------------------------")
        for c in query['choices']:
            print(c)
        print("prompt0------------------------------------------------------------------------------------")
        print(pred['prompt'][0])
        print("prompt1------------------------------------------------------------------------------------")
        print(pred['prompt'][1])
        print("gpt_output0-----------------------------------------------------------------------------------")
        print(pred['gpt_output'][0])
        print("prompt2------------------------------------------------------------------------------------")
        print(pred['prompt'][2])
        print("gpt_outpu1------------------------------------------------------------------------------------")
        print(pred['gpt_output'][1])
        break
```

```python
for query_id, query in enumerate(query_meta_data_llava):
    assert len(query['pos_candidates']) == 1
    gt_node_id = query['pos_candidates'][0]['backend_node_id']
    if gt_node_id in [choice[0] for choice in query['choices']]:
        pred = predictions_llava[query_id]
        print(gt_node_id)
        for c in query['choices']:
            print(c)
        print(pred['prompt'][2])
        print(pred['gpt_output'][0])
        print(pred['gpt_output'][1])
        break
```
